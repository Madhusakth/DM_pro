```python
import csv
import operator
import numpy as np
from PIL import Image
import os, os.path
from sklearn.model_selection import train_test_split
from collections import Counter
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
value=14951

print("Started")
csvfile = open('D:/train.csv', 'r')
csvreader = csv.reader(csvfile)
data = [line[:3] for line in csvreader]
wale=[]
for entry in data[1:]:
    wale.append(entry[2])
rows=Counter(wale).most_common(value)
res_list = [x[0] for x in rows]
x=[]
for i in res_list:
    x.append(int(i))
res_list = [x[1] for x in rows]
y=[]
for i in res_list:
    y.append(int(i))
m=sorted(range(len(x)), key=lambda k: x[k])
a=[]
a=[0]*value
b=[0]*value
for i in range(value):
    a[i]=str(i)
    b[i]=0
for i in range(len(x)):
    temp=x[i]
    b[temp]=y[i]
print("final")
temp=plt.bar(a,b)

print(max(b))
plt.xlabel('Class Label')
#plt.xticks([a[0],a[2500],a[5000],a[7500],a[10000],a[12500],a[14949]], visible=True, rotation="horizontal")
plt.ylabel('Number of Images')
plt.title('Label Distribution')
 
plt.show()
print("Done")
```
