```python
import csv
import operator
import numpy as np
from PIL import Image
import os, os.path
from sklearn.model_selection import train_test_split
from collections import Counter
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


print("Started")
csvfile = open('D:/train.csv', 'r')
csvreader = csv.reader(csvfile)
data = [line[:3] for line in csvreader]
wale=[]
mine={}

#print(wale)
for entry in data[1:]:
    wale.append(entry[2])
for i in wale:
    m=int(i)
    mine.update({m:mine.get(m, 0)+1})
print(type(mine))

miner=[]
miner=[0]*14551


for i in range(14551):
    miner[i]=mine.get(i,0)
temp=np.array(miner)
x=temp.argsort()[:10]
print("USMAN",temp)
y=[]
for i in x:
    #print(miner[i])
    y.append(miner[i])
r=[]
for i in x:
    r.append(str(i))
print(r)
print(y)
temp=plt.bar(r,y)
temp[0].set_color('red')
temp[1].set_color('green')
temp[2].set_color('blue')
temp[3].set_color('yellow')
temp[4].set_color('pink')
temp[5].set_color('brown')
temp[6].set_color('orange')
temp[7].set_color('grey')
temp[8].set_color('violet')
temp[9].set_color('black')

plt.xlabel('Class Label')
plt.ylabel('Number of Images')
plt.title('Least Frequent classes')
 
plt.show()    
print("Plotting")
```
