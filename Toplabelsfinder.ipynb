```python
import csv
import operator
import numpy as np
from PIL import Image
import os, os.path
from sklearn.model_selection import train_test_split
from collections import Counter
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


print("Started")
csvfile = open('D:/train.csv', 'r')
csvreader = csv.reader(csvfile)
data = [line[:3] for line in csvreader]
wale=[]
#print(wale)
for entry in data:
    wale.append(entry[2])
rows=Counter(wale).most_common(10)
res_list = [x[0] for x in rows]
x=[]
for i in res_list:
    x.append(int(i))
print(x)
res_list = [x[1] for x in rows]
y=[]
for i in res_list:
    y.append(int(i))
print(y)
m=sorted(range(len(x)), key=lambda k: x[k])

print(m)
a=[]
b=[]

for i in m:
    a.append(str(x[i]))
    b.append(y[i])
temp=plt.bar(a,b)
temp[0].set_color('red')
temp[1].set_color('green')
temp[2].set_color('blue')
temp[3].set_color('yellow')
temp[4].set_color('pink')
temp[5].set_color('brown')
temp[6].set_color('orange')
temp[7].set_color('grey')
temp[8].set_color('violet')
temp[9].set_color('black')

plt.xlabel('Class Label')
plt.ylabel('Number of Images')
plt.title('Top 10 Frequent classes')
 
plt.show()
```
